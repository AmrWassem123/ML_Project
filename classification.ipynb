```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split, KFold, StratifiedKFold, GridSearchCV, cross_val_score, learning_curve
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
from sklearn.feature_selection import SelectFromModel
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('output.csv').drop(['fare_amount','total_amount'],axis=1)

# Define the feature and target variables
X = df.drop('trip_price_class', axis=1)
y = df['trip_price_class']

# Instantiate the scaler
scaler = MinMaxScaler()

# Fit the scaler to the training set
scaler.fit(X)

# Transform the training and test sets
X_scaled = scaler.transform(X)

# Split the data into training and test sets.
# assume that the test data is not provided from the begining.
# so the steps for stratisfied k-folding, hyperparameter tuning 
# & fitting the model is done on the training test portion
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Define the classifiers
classifiers = {
    'Naive Bayes': MultinomialNB(),
    'Decision Tree': DecisionTreeClassifier(),
    'KNN': KNeighborsClassifier(),
    'Random Forest': RandomForestClassifier(n_estimators=50, max_depth=10, max_features='sqrt', n_jobs=-1, random_state=42),
    'Logistic Regression': LogisticRegression(solver='saga', tol=0.01)
}

# Define the hyperparameter tuning space
param_grid = {
    'Random Forest': {'n_estimators': [10, 50, 100], 'max_depth': [None, 5, 10]},
    'KNN': {
        'n_neighbors': [3,5,7],  # Number of neighbors to consider: try values 1 through 30
        'p': [1, 2]  # p=1 is Manhattan distance, p=2 is Euclidean distance
        },
    'Decision Tree': {'max_depth': [None, 5, 10]},
    'Naive Bayes': {},
    'Logistic Regression': {'C': [0.1, 1, 10]}
}

# Define the evaluation metrics
evaluation_metrics = ['accuracy', 'f1_macro', 'roc_auc']

# Define the k-fold cross-validation object
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Define the learning curve plotting function
def plot_learning_curve(estimator, X, y, cv, scoring):
    train_sizes, train_scores, test_scores = learning_curve(estimator, X, y, cv=cv, scoring=scoring)
    plt.plot(train_sizes, train_scores.mean(axis=1), label='Training Score')
    plt.plot(train_sizes, test_scores.mean(axis=1), label='Cross-validation Score')
    plt.xlabel('Training Size')
    plt.ylabel('Score')
    plt.title('Learning Curve')
    plt.legend()
    plt.show()

# Iterate over the classifiers and perform hyperparameter tuning, k-fold cross-validation, and evaluation
for name, clf in classifiers.items():
    print(f"Training {name} classifier...")
    if (name == 'KNN'):
        # Assume that you want to take a sample that is 10% of your original data
        sample_size = 0.2

        # Use train_test_split to create a stratified sample
        X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=(1-sample_size), stratify=y, random_state=42)
    else:
        X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
        if (name == 'Logistic Regression'):
            # Feature selection
            selector = SelectFromModel(estimator=LogisticRegression()).fit(X_train, y_train)
            X_train = selector.transform(X_train)
            X_test = selector.transform(X_test)

    # Perform hyperparameter tuning using GridSearchCV
    grid_search = GridSearchCV(clf, param_grid[name], cv=kfold, scoring='f1_macro')
    grid_search.fit(X_train, y_train)
    best_clf = grid_search.best_estimator_
    print(f"Best hyperparameters: {grid_search.best_params_}")
    
    # Perform k-fold cross-validation
    scores = cross_val_score(best_clf, X_train, y_train, cv=kfold, scoring='f1_macro')
    print(f"Cross-validation scores: {scores}")
    
    # Evaluate the model using confusion matrix and ROC curve
    y_pred = best_clf.predict(X_test)
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print(classification_report(y_test, y_pred))
    cm = confusion_matrix(y_test, y_pred)
    print(f"Confusion matrix:\n{cm}")
    fpr, tpr, _ = roc_curve(y_test, best_clf.predict_proba(X_test)[:, 1], pos_label=1)
    roc_auc_score = auc(fpr, tpr)
    print(f"AUC-ROC score: {roc_auc_score:.3f}")
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc_score)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Receiver Operating Characteristic - {name}')
    plt.legend(loc="lower right")
    plt.show()
    
    # Plot the learning curve
    plot_learning_curve(best_clf, X_train, y_train, kfold, 'f1_macro')
    
    print()
```

    - Uses stratified k-fold cross-validation for model validation.
